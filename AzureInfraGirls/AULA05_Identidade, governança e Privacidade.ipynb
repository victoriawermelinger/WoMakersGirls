## **Govenaça e conformidade no Azure**

### **Microsoft Purview**



Microsoft Purview é uma solução de governança de dados unificada que permite às organizações gerenciar seus dados em toda a empresa. Com ele, você pode catalogar, rastrear e gerenciar metadados e linhagem de dados em várias fontes, incluindo bancos de dados, data lakes e aplicativos SaaS.

Aqui estão algumas funcionalidades principais do Microsoft Purview:

1. **Catalogação de Dados**: Permite criar um catálogo de dados empresariais que facilita a descoberta de dados por meio de um portal centralizado.

2. **Mapeamento de Linhagem de Dados**: Rastreia a origem dos dados e como eles fluem através de diferentes sistemas, ajudando a entender o impacto das mudanças nos dados.

3. **Gestão de Políticas de Dados**: Implementa políticas de acesso e uso de dados para garantir conformidade com regulamentos e normas corporativas.

4. **Data Insights**: Fornece insights sobre os dados, incluindo a qualidade e o uso, ajudando a identificar áreas de melhoria.

5. **Integração com Outras Ferramentas**: Se integra com várias outras ferramentas e serviços de dados, como Azure Data Factory, Power BI, e SQL Server, facilitando a governança de dados em um ecossistema híbrido.

6. **Segurança e Conformidade**: Ajuda a garantir que os dados estejam protegidos e em conformidade com regulamentos como GDPR e CCPA.

Para começar a usar o Microsoft Purview, você precisará de uma assinatura do Azure e poderá configurar a solução diretamente no portal do Azure.

### **Azure Policy**



Azure Policy é um serviço no Microsoft Azure que permite criar, atribuir e gerenciar políticas que aplicam regras e efeitos aos recursos do Azure, garantindo que esses recursos estejam em conformidade com os padrões corporativos e regulatórios.

Aqui estão algumas das funcionalidades principais do Azure Policy:

1. **Criação de Políticas**: Permite criar políticas personalizadas ou usar políticas predefinidas para gerenciar os recursos do Azure. As políticas definem regras que podem ser aplicadas aos recursos, como requisitos de tagging, restrições de localização e tipos de recursos permitidos.

2. **Atribuição de Políticas**: As políticas podem ser atribuídas a escopos específicos, como grupos de recursos, assinaturas ou grupos de gerenciamento, permitindo uma aplicação flexível das regras.

3. **Avaliação de Conformidade**: Azure Policy avalia continuamente os recursos em relação às políticas atribuídas, fornecendo insights sobre a conformidade e destacando os recursos que não estão em conformidade.

4. **Ação Remediativa**: Algumas políticas podem ser configuradas para executar ações remediativas automaticamente, corrigindo recursos não conformes, como adicionar tags ausentes ou mover recursos para regiões permitidas.

5. **Gestão de Iniciativas**: Permite agrupar várias políticas em uma iniciativa, facilitando a gestão de um conjunto de políticas relacionadas como uma unidade única.

6. **Integração com Azure Blueprints**: Azure Policy pode ser usado em conjunto com Azure Blueprints para criar padrões de governança e conformidade que podem ser facilmente aplicados a novas implementações de ambientes no Azure.

7. **Auditoria e Relatórios**: Oferece ferramentas de auditoria e relatórios para monitorar e revisar a conformidade dos recursos, ajudando a garantir que as políticas estejam sendo seguidas e a identificar áreas de melhoria.

Para começar a usar o Azure Policy, você pode acessar o portal do Azure, navegar até o serviço Azure Policy e começar a criar e atribuir políticas conforme necessário para garantir a conformidade e a governança de seus recursos no Azure.



Para bloquear um recurso no Azure, siga os passos abaixo conforme a captura de tela fornecida:

1. **Navegue até o Recurso**: Acesse o recurso específico que deseja bloquear no portal do Azure.

2. **Acesse a Opção de Bloqueios**:
   - No menu à esquerda, selecione a opção "Bloqueios" (marcado como passo 2 na imagem).

3. **Adicionar um Novo Bloqueio**:
   - Clique no botão "+ Adicionar" (marcado como passo 3 na imagem).

4. **Configurar o Bloqueio**:
   - Preencha os detalhes do bloqueio na janela "Adicionar bloqueio":
     - **Nome do bloqueio**: Insira um nome descritivo para o bloqueio.
     - **Tipo de bloqueio**: Selecione o tipo de bloqueio. Existem dois tipos principais:
       - **Somente leitura** (`ReadOnly`): Permite apenas operações de leitura no recurso. Não é possível excluir ou modificar o recurso.
       - **Excluir** (`Delete`): Bloqueia a exclusão do recurso, mas permite modificações.
     - **Observações**: Insira quaisquer observações adicionais sobre o bloqueio.

5. **Salvar o Bloqueio**:
   - Clique em "OK" para salvar e aplicar o bloqueio ao recurso.

Esses passos ajudarão a proteger seus recursos contra alterações acidentais ou exclusões, garantindo uma maior segurança e conformidade com as políticas de governança de dados.

### Portal de Confiança do Serviço 



O **Portal de Confiança do Serviço** (Service Trust Portal) da Microsoft é uma plataforma onde a empresa oferece informações detalhadas sobre como os serviços da Microsoft implementam a segurança, privacidade e conformidade. Este portal é essencial para organizações que desejam entender as práticas de segurança e conformidade da Microsoft e como elas podem ajudar a atender às suas próprias exigências regulatórias e corporativas.

### Funcionalidades Principais do Portal de Confiança do Serviço:

1. **Documentação de Conformidade**:
   - Acesso a relatórios de auditoria, certificações e outros documentos que demonstram a conformidade dos serviços da Microsoft com padrões e regulamentos como ISO, SOC, GDPR, e HIPAA.

2. **Guias de Implementação**:
   - Fornece guias e práticas recomendadas para a implementação segura e em conformidade dos serviços da Microsoft.

3. **Centro de Conformidade**:
   - Um repositório centralizado de informações e recursos sobre conformidade, que ajuda as organizações a entender como os serviços da Microsoft podem atender aos requisitos de conformidade específicos.

4. **Ferramentas de Avaliação**:
   - Ferramentas para avaliar e gerenciar a conformidade e os riscos de segurança, incluindo a Ferramenta de Gerenciamento de Conformidade (Compliance Manager).

5. **Transparência de Dados**:
   - Informações detalhadas sobre onde os dados estão armazenados, quem pode acessá-los e como eles são protegidos.

6. **Notícias e Atualizações**:
   - Atualizações sobre novos desenvolvimentos, certificações e mudanças nas práticas de segurança e conformidade da Microsoft.

### Como Acessar o Portal de Confiança do Serviço:

1. **Visite o Site**:
   - Acesse o Portal de Confiança do Serviço através do link: [Service Trust Portal](https://servicetrust.microsoft.com/).

2. **Login**:
   - Faça login com sua conta da Microsoft para acessar documentos e recursos específicos.

3. **Navegação e Pesquisa**:
   - Utilize a barra de pesquisa e os menus de navegação para encontrar informações e documentos específicos sobre conformidade, segurança e privacidade.

4. **Downloads**:
   - Baixe relatórios de auditoria, certificações e outros documentos relevantes para seu próprio uso e para compartilhar com stakeholders.

O Portal de Confiança do Serviço é uma ferramenta essencial para qualquer organização que utilize serviços da Microsoft e precise garantir que suas práticas de segurança e conformidade estejam em linha com os padrões globais.

# **Gernciamento de Custos no Azure**



A imagem descreve os fatores que podem afetar os custos no Azure, destacando a transição de CapEx (despesa de capital) para OpEx (despesa operacional) ao utilizar a infraestrutura de nuvem da Microsoft. Aqui estão os principais pontos abordados:

### Fatores que Podem Afetar os Custos no Azure:

1. **Tipo de Recurso**:
   - O tipo de recurso utilizado (como máquinas virtuais, armazenamento, bases de dados, etc.) tem um impacto significativo nos custos.

2. **Consumo**:
   - O nível de utilização dos recursos, como tempo de execução, quantidade de dados processados ou armazenados, afeta diretamente os custos operacionais.

3. **Manutenção**:
   - Custos associados à manutenção dos recursos, incluindo atualizações e suporte técnico.

4. **Geografia do App Selecionado**:
   - A localização geográfica dos recursos influencia os custos devido a variações nos preços regionais e na latência de rede.

5. **Tipo de Assinatura**:
   - O tipo de assinatura ou plano escolhido pode alterar os custos. Diferentes planos oferecem benefícios e preços variados.

6. **Azure Marketplace**:
   - Utilizar soluções e serviços adicionais disponíveis no Azure Marketplace pode adicionar custos extras dependendo dos serviços escolhidos.

### Detalhes na Imagem:

- A imagem mostra a interface de criação de uma máquina virtual no portal do Azure, onde é possível configurar vários parâmetros que afetam o custo, como:
  - Localização geográfica
  - Tipo e tamanho da máquina virtual
  - Configurações de armazenamento e rede

- Há também uma seção destacando o armazenamento de blobs, com opções para configurar a replicação e o nível de acesso aos dados, que também impactam os custos.

### Importância da Gestão de Custos:

Gerenciar os custos no Azure é crucial para otimizar os gastos e garantir que a infraestrutura da nuvem seja utilizada de forma eficiente. Utilizar ferramentas de monitoramento e gestão de custos do Azure pode ajudar a acompanhar e controlar os gastos, garantindo que as operações permaneçam dentro do orçamento planejado.
