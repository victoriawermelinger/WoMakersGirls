### **Microsoft Purview**



Microsoft Purview  uma soluo de governana de dados unificada que permite s organizaes gerenciar seus dados em toda a empresa. Com ele, voc pode catalogar, rastrear e gerenciar metadados e linhagem de dados em vrias fontes, incluindo bancos de dados, data lakes e aplicativos SaaS.

Aqui esto algumas funcionalidades principais do Microsoft Purview:

1. **Catalogao de Dados**: Permite criar um catlogo de dados empresariais que facilita a descoberta de dados por meio de um portal centralizado.

2. **Mapeamento de Linhagem de Dados**: Rastreia a origem dos dados e como eles fluem atravs de diferentes sistemas, ajudando a entender o impacto das mudanas nos dados.

3. **Gesto de Polticas de Dados**: Implementa polticas de acesso e uso de dados para garantir conformidade com regulamentos e normas corporativas.

4. **Data Insights**: Fornece insights sobre os dados, incluindo a qualidade e o uso, ajudando a identificar reas de melhoria.

5. **Integrao com Outras Ferramentas**: Se integra com vrias outras ferramentas e servios de dados, como Azure Data Factory, Power BI, e SQL Server, facilitando a governana de dados em um ecossistema hbrido.

6. **Segurana e Conformidade**: Ajuda a garantir que os dados estejam protegidos e em conformidade com regulamentos como GDPR e CCPA.

Para comear a usar o Microsoft Purview, voc precisar de uma assinatura do Azure e poder configurar a soluo diretamente no portal do Azure.

### **Azure Policy**



Azure Policy  um servio no Microsoft Azure que permite criar, atribuir e gerenciar polticas que aplicam regras e efeitos aos recursos do Azure, garantindo que esses recursos estejam em conformidade com os padres corporativos e regulatrios.

Aqui esto algumas das funcionalidades principais do Azure Policy:

1. **Criao de Polticas**: Permite criar polticas personalizadas ou usar polticas predefinidas para gerenciar os recursos do Azure. As polticas definem regras que podem ser aplicadas aos recursos, como requisitos de tagging, restries de localizao e tipos de recursos permitidos.

2. **Atribuio de Polticas**: As polticas podem ser atribudas a escopos especficos, como grupos de recursos, assinaturas ou grupos de gerenciamento, permitindo uma aplicao flexvel das regras.

3. **Avaliao de Conformidade**: Azure Policy avalia continuamente os recursos em relao s polticas atribudas, fornecendo insights sobre a conformidade e destacando os recursos que no esto em conformidade.

4. **Ao Remediativa**: Algumas polticas podem ser configuradas para executar aes remediativas automaticamente, corrigindo recursos no conformes, como adicionar tags ausentes ou mover recursos para regies permitidas.

5. **Gesto de Iniciativas**: Permite agrupar vrias polticas em uma iniciativa, facilitando a gesto de um conjunto de polticas relacionadas como uma unidade nica.

6. **Integrao com Azure Blueprints**: Azure Policy pode ser usado em conjunto com Azure Blueprints para criar padres de governana e conformidade que podem ser facilmente aplicados a novas implementaes de ambientes no Azure.

7. **Auditoria e Relatrios**: Oferece ferramentas de auditoria e relatrios para monitorar e revisar a conformidade dos recursos, ajudando a garantir que as polticas estejam sendo seguidas e a identificar reas de melhoria.

Para comear a usar o Azure Policy, voc pode acessar o portal do Azure, navegar at o servio Azure Policy e comear a criar e atribuir polticas conforme necessrio para garantir a conformidade e a governana de seus recursos no Azure.



Para bloquear um recurso no Azure, siga os passos abaixo conforme a captura de tela fornecida:

1. **Navegue at o Recurso**: Acesse o recurso especfico que deseja bloquear no portal do Azure.

2. **Acesse a Opo de Bloqueios**:
   - No menu  esquerda, selecione a opo "Bloqueios" (marcado como passo 2 na imagem).

3. **Adicionar um Novo Bloqueio**:
   - Clique no boto "+ Adicionar" (marcado como passo 3 na imagem).

4. **Configurar o Bloqueio**:
   - Preencha os detalhes do bloqueio na janela "Adicionar bloqueio":
     - **Nome do bloqueio**: Insira um nome descritivo para o bloqueio.
     - **Tipo de bloqueio**: Selecione o tipo de bloqueio. Existem dois tipos principais:
       - **Somente leitura** (`ReadOnly`): Permite apenas operaes de leitura no recurso. No  possvel excluir ou modificar o recurso.
       - **Excluir** (`Delete`): Bloqueia a excluso do recurso, mas permite modificaes.
     - **Observaes**: Insira quaisquer observaes adicionais sobre o bloqueio.

5. **Salvar o Bloqueio**:
   - Clique em "OK" para salvar e aplicar o bloqueio ao recurso.

Esses passos ajudaro a proteger seus recursos contra alteraes acidentais ou excluses, garantindo uma maior segurana e conformidade com as polticas de governana de dados.

3. **Manuteno**:
   - Custos associados  manuteno dos recursos, incluindo atualizaes e suporte tcnico.

4. **Geografia do App Selecionado**:
   - A localizao geogrfica dos recursos influencia os custos devido a variaes nos preos regionais e na latncia de rede.

5. **Tipo de Assinatura**:
   - O tipo de assinatura ou plano escolhido pode alterar os custos. Diferentes planos oferecem benefcios e preos variados.

6. **Azure Marketplace**:
   - Utilizar solues e servios adicionais disponveis no Azure Marketplace pode adicionar custos extras dependendo dos servios escolhidos.

### **Calculadoras no Azure** 


O Azure oferece diversas ferramentas de clculo para ajudar os usurios a estimar e gerenciar os custos dos servios e recursos na plataforma. Aqui esto as principais calculadoras disponveis no Azure:

### 1. **Calculadora de Preos do Azure**
A Calculadora de Preos do Azure  uma ferramenta essencial que permite estimar os custos dos servios e recursos do Azure antes de implement-los. Com ela, voc pode:

### 2. **Azure TCO (Total Cost of Ownership) Calculator**

- Inserir detalhes sobre sua infraestrutura atual, como servidores, armazenamento, custos de energia, e mo de obra.
- Estimar os custos de migrao para o Azure e compar-los com os custos atuais.
- Visualizar a economia potencial ao mover para a nuvem do Azure.

**Acesse a Calculadora de TCO do Azure**: [Calculadora de TCO do Azure](https://azure.microsoft.com/pt-br/pricing/tco/calculator/)

### 3. **Azure Cost Management and Billing**
Embora no seja uma calculadora tradicional, o Azure Cost Management and Billing  uma ferramenta poderosa para monitorar, alocar e otimizar os custos dos servios do Azure. Com ele, voc pode:

- Visualizar os gastos atuais e histricos.
- Configurar oramentos e alertas para evitar gastos excessivos.
- Analisar o consumo de recursos e identificar reas para otimizao de custos.
- Alocar custos por departamento, projeto ou equipe para uma melhor gesto financeira.

**Acesse o Azure Cost Management and Billing**: [Azure Cost Management and Billing](https://azure.microsoft.com/pt-br/services/cost-management/)

### 4. **Calculadora de Economias do Azure Hybrid Benefit**
Esta calculadora ajuda a entender como o Azure Hybrid Benefit pode reduzir os custos ao usar suas licenas existentes do Windows Server e SQL Server no Azure.

**Acesse a Calculadora de Economias do Azure Hybrid Benefit**: [Azure Hybrid Benefit](https://azure.microsoft.com/pt-br/pricing/hybrid-benefit/)

### **Gerenciamento de custos da Microsoft**


O gerenciamento de custos da Microsoft Azure envolve um conjunto de ferramentas e prticas para ajudar as organizaes a monitorar, alocar e otimizar os gastos na nuvem. A seguir esto os principais componentes e estratgias para o gerenciamento de custos no Azure:

### Ferramentas Principais para Gerenciamento de Custos no Azure

1. **Azure Cost Management and Billing**
   - **Descrio**: Uma ferramenta integrada que permite monitorar, alocar e otimizar os custos do Azure.
   - **Funcionalidades**:
     - **Relatrios de Gastos**: Visualizao de gastos atuais e histricos com grficos e tabelas detalhadas.
     - **Oramentos e Alertas**: Configurao de oramentos para recursos especficos e alertas para evitar gastos excessivos.
     - **Anlise de Consumo**: Identificao de padres de uso e reas onde os custos podem ser reduzidos.
     - **Alocao de Custos**: Distribuio de custos por departamento, projeto ou equipe para uma melhor gesto financeira.

2. **Azure Pricing Calculator**
   - **Descrio**: Uma ferramenta para estimar os custos de diferentes servios e recursos do Azure antes da implementao.
   - **Funcionalidades**:
     - **Simulao de Custos**: Configurao de recursos e servios para obter uma estimativa de custo detalhada.
     - **Comparao de Configuraes**: Comparao de diferentes cenrios de implementao para encontrar a opo mais econmica.

3. **Azure Total Cost of Ownership (TCO) Calculator**
   - **Descrio**: Uma ferramenta que ajuda a comparar os custos totais de propriedade da infraestrutura local versus os custos de migrar para o Azure.
   - **Funcionalidades**:
     - **Comparao de Custos**: Insero de detalhes da infraestrutura atual e comparao com os custos estimados no Azure.
     - **Visualizao de Economia**: Anlise da economia potencial ao migrar para a nuvem do Azure.

4. **Azure Advisor**
   - **Descrio**: Um servio que fornece recomendaes personalizadas para otimizar os recursos do Azure.
   - **Funcionalidades**:
     - **Recomendaes de Custo**: Sugestes para reduzir gastos, como redimensionar ou desligar recursos subutilizados.
     - **Otimizao de Recursos**: Identificao de oportunidades para melhorar a performance e eficincia dos recursos.

### Estratgias para Otimizar Custos no Azure

1. **Dimensionamento Correto dos Recursos**
   - Assegure-se de que os recursos, como VMs e bases de dados, estejam dimensionados corretamente para evitar pagar por capacidade no utilizada.

2. **Uso de Instncias Reservadas**
   - Considere adquirir instncias reservadas para cargas de trabalho estveis e previsveis, o que pode resultar em economias significativas em comparao com o modelo de pagamento conforme o uso.

3. **Desligamento de Recursos Ociosos**
   - Implemente polticas e automaes para desligar recursos que no esto sendo utilizados, como VMs e ambientes de desenvolvimento/teste fora do horrio de trabalho.

4. **Adoo de Arquiteturas Serverless**
   - Utilize servios serverless, como Azure Functions e Azure Logic Apps, para cargas de trabalho que podem se beneficiar de uma escalabilidade automtica e um modelo de pagamento baseado no consumo real.

5. **Monitoramento Contnuo e Ajuste**
   - Utilize ferramentas como o Azure Monitor e o Azure Cost Management para monitorar continuamente o uso e os custos, ajustando conforme necessrio para otimizar os gastos.

6. **Treinamento e Governana**
   - Capacite a equipe com o conhecimento e as melhores prticas para a gesto eficiente dos recursos na nuvem. Estabelea polticas de governana claras para o uso e a alocao de recursos.

### Concluso

O gerenciamento eficaz dos custos no Azure envolve a utilizao de vrias ferramentas e estratgias para monitorar, alocar e otimizar os gastos. Ao adotar uma abordagem proativa e utilizar as ferramentas disponveis, as organizaes podem maximizar o valor obtido dos seus investimentos na nuvem do Azure, garantindo ao mesmo tempo que os recursos sejam utilizados de maneira eficiente e econmica.
