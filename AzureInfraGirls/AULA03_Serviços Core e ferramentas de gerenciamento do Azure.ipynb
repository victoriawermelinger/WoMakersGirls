# Servios Core e ferramentas de gerenciamento do Azure 

Os servios principais e as ferramentas de gerenciamento do Azure so essenciais para administrar recursos na nuvem de forma eficaz. Aqui esto alguns dos principais servios e ferramentas:

### Servios Core do Azure

1. **Computao**
   - **Mquinas Virtuais (VMs)**: Infraestrutura como Servio (IaaS) para executar cargas de trabalho na nuvem.
   - **Azure Kubernetes Service (AKS)**: Orquestrao de contineres para gerenciar aplicativos em contineres.
   - **Azure Functions**: Plataforma de Computao sem Servidor (FaaS) para executar cdigo sob demanda.

2. **Armazenamento**
   - **Azure Blob Storage**: Armazenamento de objetos escalvel para dados no estruturados.
   - **Azure File Storage**: Armazenamento de arquivos compatvel com SMB.
   - **Azure Disk Storage**: Discos gerenciados para VMs.

3. **Banco de Dados**
   - **Azure SQL Database**: Banco de dados relacional como servio (DBaaS).
   - **Azure Cosmos DB**: Banco de dados NoSQL globalmente distribudo.
   - **Azure Database for MySQL/PostgreSQL/MariaDB**: Bancos de dados relacionais gerenciados.

4. **Rede**
   - **Azure Virtual Network (VNet)**: Rede privada na nuvem.
   - **Azure Load Balancer**: Balanceamento de carga de trfego para alta disponibilidade.
   - **Azure Application Gateway**: Gerenciamento de trfego com balanceamento de carga e firewall de aplicativos web.

5. **Identidade e Acesso**
   - **Azure Active Directory (AAD)**: Gerenciamento de identidades e acessos.
   - **Azure Multi-Factor Authentication (MFA)**: Segurana adicional com autenticao multifator.

### Ferramentas de Gerenciamento do Azure

1. **Azure Portal**
   - Interface web grfica para gerenciar e configurar servios do Azure.

2. **Azure PowerShell**
   - Linha de comando para automatizao e gerenciamento de recursos Azure.

3. **Azure CLI**
   - Ferramenta de linha de comando multiplataforma para gerenciamento de recursos.

4. **Azure Resource Manager (ARM)**
   - Servio de gerenciamento para provisionamento de recursos de maneira declarativa via templates.

5. **Azure Monitor**
   - Servio de monitoramento para coleta e anlise de mtricas e logs.

6. **Azure Advisor**
   - Servio que fornece recomendaes personalizadas para otimizao de recursos Azure.

7. **Azure Security Center**
   - Ferramenta para gesto e monitoramento da segurana de recursos Azure.

8. **Azure Cost Management**
   - Servio para monitoramento e controle de custos e otimizao de gastos.

9. **Azure DevOps**
   - Conjunto de ferramentas para desenvolvimento, entrega e manuteno de software (CI/CD).

Esses servios e ferramentas formam a base para a criao, implantao, monitoramento e gerenciamento de solues no Azure, permitindo uma abordagem abrangente para gerenciar a infraestrutura de TI na nuvem.





# Ferramentas para interagir com Azure

Interagir com o Azure pode ser feito por meio de vrias ferramentas, cada uma oferecendo diferentes funcionalidades para administrar, configurar e gerenciar recursos na nuvem. Aqui esto as principais ferramentas para interagir com o Azure:

1. **Azure CLI**
   - Uma interface de linha de comando multiplataforma que permite gerenciar os recursos do Azure. Pode ser usada em sistemas operacionais Windows, macOS e Linux.
   - Comandos comuns incluem `az login`, `az group create`, `az vm create`, entre outros.

2. **Azure PowerShell**
   - Um conjunto de mdulos do PowerShell que permitem gerenciar o Azure com cmdlets. Ideal para automatizao e script de operaes administrativas.
   - Exemplos de cmdlets: `Connect-AzAccount`, `New-AzResourceGroup`, `New-AzVM`.

1. **Azure Portal**
   - Interface web grfica acessvel via navegador que permite gerenciar e configurar todos os servios do Azure.
   - URL: [Azure Portal](https://portal.azure.com)

2. **Azure Mobile App**
   - Aplicativo mvel disponvel para Android e iOS que permite monitorar e gerenciar recursos do Azure em movimento.

### Ferramentas de Desenvolvimento

1. **Azure Resource Manager (ARM) Templates**
   - Arquivos JSON que definem a infraestrutura e a configurao dos recursos do Azure de maneira declarativa. Usados para automao e provisionamento de recursos de forma consistente.

2. **Terraform**
   - Ferramenta de Infraestrutura como Cdigo (IaC) que permite definir e provisionar infraestrutura do Azure utilizando a linguagem de configurao HCL (HashiCorp Configuration Language).

### Integrao e Automao

1. **Azure DevOps**
   - Conjunto de ferramentas para planejamento de projetos, desenvolvimento de cdigo, entrega contnua e monitoramento de aplicaes. Inclui Azure Repos, Azure Pipelines, Azure Boards, entre outros.

2. **GitHub Actions**

### Monitoramento e Anlise

1. **Azure Monitor**
   - Servio de monitoramento que coleta, analisa e age sobre dados de telemetria de recursos do Azure. Inclui mtricas, logs, alertas e dashboards personalizados.

2. **Log Analytics**
   - Parte do Azure Monitor, permite consultar e analisar logs de recursos do Azure utilizando a linguagem de consulta Kusto (KQL).

### Ferramentas de Gerenciamento de Custos

1. **Azure Cost Management and Billing**

### Outras Ferramentas teis

   - Ferramenta standalone para gerenciar dados no armazenamento do Azure, como blobs, filas, tabelas e arquivos.

2. **Azure Data Studio**

### SDKs e APIs

Essas ferramentas proporcionam flexibilidade e capacidade para interagir com o Azure de maneiras que melhor se adaptem s necessidades especficas de desenvolvedores, administradores de sistemas e arquitetos de solues.



## **Azure Arc**
Azure Arc  uma plataforma da Microsoft que permite estender os servios e a gesto do Azure a qualquer infraestrutura, seja on-premises, em outras nuvens ou em data centers de borda. Com o Azure Arc, as organizaes podem gerenciar recursos espalhados em ambientes hbridos e multicloud como se estivessem no Azure. Aqui esto os principais componentes e funcionalidades do Azure Arc:

### Componentes Principais do Azure Arc

1. **Azure Arc para Servidores**
   - Permite gerenciar servidores fsicos e mquinas virtuais que esto fora do Azure (on-premises ou em outras nuvens) como se fossem recursos nativos do Azure.
   - Integrao com Azure Policy, Azure Monitor e Azure Security Center para governana e monitoramento centralizados.

2. **Azure Arc para Kubernetes**
   - Gerenciamento centralizado de clusters Kubernetes, seja no Azure, on-premises ou em outras nuvens.
   - Facilita a implantao de aplicativos e a configurao de polticas uniformes para todos os clusters.

3. **Azure Arc para Aplicaes de Dados**
   - Permite executar servios de dados do Azure, como Azure SQL Managed Instance e PostgreSQL Hyperscale, em qualquer infraestrutura.
   - Proporciona uma experincia consistente de gerenciamento e governana de dados em ambientes hbridos.

### Funcionalidades do Azure Arc

1. **Gesto Unificada**
   - Uma nica interface de gerenciamento para recursos espalhados em diferentes ambientes, utilizando o Azure Portal, Azure CLI, e Azure PowerShell.

2. **Governana e Conformidade**
   - Aplicao de polticas e compliance de maneira uniforme em toda a infraestrutura usando Azure Policy.
   - Monitoramento de segurana e conformidade com Azure Security Center.

3. **Automao e DevOps**
   - Integrao com ferramentas de DevOps como Azure DevOps e GitHub Actions para CI/CD, automao de infraestrutura e configurao.

4. **Monitoramento e Insights**
   - Coleta e anlise de logs e mtricas de todos os recursos gerenciados, proporcionando visibilidade abrangente e insights acionveis atravs do Azure Monitor.

5. **Gesto de Identidades**
   - Integrao com Azure Active Directory para gerenciamento de identidades e controle de acesso unificado em toda a infraestrutura.

### Casos de Uso do Azure Arc

1. **Ambientes Hbridos**
   - Empresas que possuem uma combinao de recursos on-premises e na nuvem podem usar o Azure Arc para ter um controle centralizado e polticas uniformes.

2. **Multicloud**
   - Organizaes que utilizam mltiplas nuvens (Azure, AWS, Google Cloud) podem gerenciar todos esses recursos atravs de uma nica interface com Azure Arc.

3. **Aplicaes de Dados**
   - Executar servios de dados do Azure em data centers locais para atender a requisitos de latncia, conformidade e residncia de dados, mantendo uma experincia de gesto consistente com o Azure.

4. **Edge Computing**
   - Gerenciar dispositivos e servios de borda que esto fora dos data centers tradicionais, aplicando polticas e segurana de forma centralizada.

### Benefcios do Azure Arc

- **Consistncia**: Fornece uma abordagem unificada para gesto e governana de recursos, independentemente de onde estejam localizados.
- **Eficincia Operacional**: Reduz a complexidade operacional atravs de ferramentas e processos padronizados.
- **Segurana e Conformidade**: Melhora a segurana e conformidade com polticas e controles centralizados.
- **Flexibilidade**: Permite que as organizaes escolham a infraestrutura que melhor atende s suas necessidades, sem sacrificar a capacidade de gerenciamento.

   


Azure Arc  uma soluo poderosa para organizaes que buscam simplificar e otimizar a gesto de recursos em ambientes hbridos e multicloud, proporcionando uma experincia de gerenciamento e governana coesa e integrada com o Azure.

Azure Resource Manager (ARM)  o servio de gerenciamento de implantao e operaes do Azure. Ele fornece uma camada de gerenciamento que permite criar, atualizar e excluir recursos em sua conta do Azure. ARM oferece vrios benefcios e recursos para gerenciar recursos de forma eficiente e organizada. Aqui esto os principais aspectos e funcionalidades do Azure Resource Manager:

### Principais Funcionalidades do Azure Resource Manager

1. **Modelo de Implantao Consistente**
   - ARM fornece um modelo de implantao consistente e uniforme para todos os recursos do Azure. Voc pode gerenciar recursos em um grupo de recursos como uma unidade lgica.

2. **Templates do ARM**
   - Permite definir a infraestrutura como cdigo (IaC) usando templates JSON. Isso facilita a automao e a replicao de implantaes.
   - Exemplos de arquivos JSON incluem definies para mquinas virtuais, redes virtuais, contas de armazenamento, entre outros.

3. **Grupos de Recursos**
   - Um grupo de recursos  um continer lgico onde voc gerencia e organiza recursos do Azure. Todos os recursos em um grupo de recursos compartilham o mesmo ciclo de vida e podem ser gerenciados, provisionados e monitorados juntos.

4. **Controle de Acesso Baseado em Funo (RBAC)**
   - ARM oferece controle de acesso detalhado para recursos com RBAC. Voc pode atribuir permisses especficas a usurios, grupos e servios para garantir a segurana e a governana.

5. **Tagging de Recursos**
   - Permite adicionar tags (pares chave-valor) aos recursos para facilitar a organizao, categorizao e gerenciamento, como por departamento, ambiente ou projeto.

6. **Polticas do Azure**
   - Use Azure Policy para criar, atribuir e gerenciar polticas que aplicam regras e efeitos aos recursos. Isso ajuda a garantir que os recursos estejam em conformidade com os padres corporativos e requisitos regulatrios.

7. **Monitoramento e Alertas**
   - ARM integra-se ao Azure Monitor para fornecer monitoramento contnuo e alertas para recursos gerenciados. Isso permite a deteco e resposta a problemas em tempo real.

### Benefcios do Azure Resource Manager

1. **Gesto Centralizada**
   - Fornece uma viso centralizada e unificada para gerenciar todos os recursos do Azure, permitindo uma administrao mais eficaz e organizada.

2. **Automao e Eficincia**
   - Os templates do ARM permitem automao e consistncia nas implantaes, reduzindo erros manuais e aumentando a eficincia operacional.

3. **Segurana e Conformidade**
   - Com RBAC e polticas do Azure, voc pode garantir que os recursos sejam acessados e configurados de acordo com as polticas de segurana e conformidade da organizao.

4. **Facilidade de Escalabilidade**
   - ARM facilita a escalabilidade horizontal e vertical de recursos, permitindo que voc adapte sua infraestrutura s necessidades de negcios de forma rpida e eficaz.

5. **Custo-efetividade**
   - Tags e polticas de governana ajudam a controlar e otimizar os custos, facilitando a atribuio de despesas e a identificao de oportunidades de economia.

### Exemplo de Template ARM

Aqui est um exemplo simples de um template ARM que cria uma mquina virtual:

**Infraestrutura como Cdigo(laaC)**




Infraestrutura como Cdigo (IaC, ou "Infrastructure as Code" em ingls)  uma prtica de gerenciamento de infraestrutura que envolve a definio e provisionamento de recursos de TI atravs de cdigo, em vez de processos manuais. Isso permite que a infraestrutura seja configurada e gerenciada de forma consistente, automatizada e reproduzvel. Aqui esto os principais conceitos, benefcios e ferramentas associadas  IaC:

### Conceitos Principais

1. **Declarao vs. Imperativa**
   - **Declarao**: Voc descreve o estado desejado da infraestrutura (por exemplo, "quero uma VM com 2 CPUs e 4GB de RAM") e a ferramenta IaC converte isso em aes concretas.
   - **Imperativa**: Voc descreve as etapas necessrias para alcanar o estado desejado (por exemplo, "crie uma VM, adicione CPUs, adicione RAM").

2. **Reusabilidade e Modularidade**
   - IaC permite a criao de componentes reutilizveis e modulares que podem ser combinados e reutilizados em diferentes partes da infraestrutura.

3. **Controle de Verso**
   - Os arquivos de configurao de IaC podem ser armazenados em sistemas de controle de verso (como Git), permitindo histrico de mudanas, colaborao e auditoria.

### Benefcios da IaC

1. **Automao e Eficincia**
   - Reduz a necessidade de configurao manual, acelerando o provisionamento e reduzindo erros humanos.

2. **Consistncia e Reprodutibilidade**
   - Garante que a infraestrutura seja configurada de forma consistente em diferentes ambientes (desenvolvimento, teste, produo).

3. **Documentao Automtica**
   - O cdigo IaC serve como documentao viva da infraestrutura, facilitando a manuteno e a transferncia de conhecimento.

4. **Escalabilidade**
   - Facilita a escalabilidade horizontal (adicionando mais instncias de um recurso) e vertical (aumentando os recursos de uma instncia).

5. **Colaborao**
   - Com o cdigo IaC armazenado em sistemas de controle de verso, vrias equipes podem colaborar de forma eficaz, revisando e aprovando mudanas na infraestrutura.

### Ferramentas Populares de IaC

1. **Terraform**
   - Uma ferramenta de cdigo aberto desenvolvida pela HashiCorp que permite definir infraestrutura em vrios provedores de nuvem (como AWS, Azure, GCP) usando a linguagem HCL (HashiCorp Configuration Language).
   - Exemplo de cdigo Terraform:
     ```hcl
     provider "azurerm" {
       features {}
     }

     resource "azurerm_resource_group" "example" {
       name     = "example-resources"
       location = "West Europe"
     }

     resource "azurerm_virtual_network" "example" {
       name                = "example-network"
       address_space       = ["10.0.0.0/16"]
       location            = azurerm_resource_group.example.location
       resource_group_name = azurerm_resource_group.example.name
     }
     ```

2. **Azure Resource Manager (ARM) Templates**
   - Arquivos JSON que definem a infraestrutura e configurao de recursos no Azure de maneira declarativa.
   - Exemplo de template ARM j foi fornecido anteriormente.

3. **Ansible**
   - Ferramenta de automao que utiliza YAML para definir tarefas e playbooks. Embora seja mais conhecida por automao de configurao, tambm pode ser usada para provisionamento de infraestrutura.
   - Exemplo de playbook Ansible:
     ```yaml

       hosts: localhost
       tasks:
         - name: Create a resource group

             name: myResourceGroup
     ```

4. **Pulumi**
   - Ferramenta que permite definir infraestrutura usando linguagens de programao convencionais como TypeScript, Python, Go, e C#.
   - Exemplo de cdigo Pulumi:
     ```typescript

     const resourceGroup = new azure.core.ResourceGroup("resourceGroup", {
       location: "WestUS",

     ```

5. **Chef e Puppet**
   - Ferramentas de automao de configurao que tambm suportam a definio de infraestrutura como cdigo. Utilizam suas prprias DSLs (Domain-Specific Languages) para descrever o estado desejado da infraestrutura.

### Prticas Recomendadas

1. **Controle de Verso**
   - Mantenha todos os arquivos de configurao em um sistema de controle de verso para rastrear mudanas e facilitar a colaborao.

2. **Modularizao**

**Modelo do ARM (ARM(Azure Resource Manager)Templates)**



Azure Resource Manager (ARM) Templates so arquivos JSON que definem a infraestrutura e a configurao dos recursos do Azure. Eles permitem que voc implemente, atualize ou exclua recursos de maneira declarativa e reproduzvel.

### Estrutura Bsica de um Template ARM

Um template ARM tpico contm as seguintes sees principais:

1. **$schema**: Especifica o esquema JSON que valida o template.
2. **contentVersion**: Especifica a verso do template.
3. **parameters**: Define os parmetros que sero passados ao template durante a implantao.
4. **variables**: Define variveis usadas no template para simplificar a definio de recursos.
5. **resources**: Lista os recursos que sero criados ou gerenciados pelo template.
6. **outputs**: Define os valores de sada que podem ser retornados aps a implantao do template.

### Exemplo de Template ARM

A seguir, um exemplo simples de um template ARM que cria um grupo de recursos e uma mquina virtual:

```json
{
  "$schema": "https://schema.management.azure.com/schemas/2019-04-01/deploymentTemplate.json#",
  "contentVersion": "1.0.0.0",
  "parameters": {
    "vmName": {
      "type": "string",
      "defaultValue": "myVM",
      "metadata": {
        "description": "Nome da Mquina Virtual"
      }
    },
    "adminUsername": {
      "type": "string",
      "metadata": {
        "description": "Nome do Administrador"
      }
    },
    "adminPassword": {
      "type": "securestring",
      "metadata": {
        "description": "Senha do Administrador"
      }
    }
  },
  "variables": {
    "location": "[resourceGroup().location]"
  },
  "resources": [
    {
      "type": "Microsoft.Compute/virtualMachines",
      "apiVersion": "2021-03-01",
      "name": "[parameters('vmName')]",
      "location": "[variables('location')]",
      "properties": {
        "hardwareProfile": {
          "vmSize": "Standard_DS1_v2"
        },
        "osProfile": {
          "computerName": "[parameters('vmName')]",
          "adminUsername": "[parameters('adminUsername')]",
          "adminPassword": "[parameters('adminPassword')]"
        },
        "storageProfile": {
          "imageReference": {
            "publisher": "Canonical",
            "offer": "UbuntuServer",
            "sku": "18.04-LTS",
            "version": "latest"
          },
          "osDisk": {
            "createOption": "FromImage"
          }
        },
        "networkProfile": {
          "networkInterfaces": [
            {
              "id": "[resourceId('Microsoft.Network/networkInterfaces', 'myNic')]"
            }
          ]
        }
      }
    }
  ],
  "outputs": {
    "vmId": {
      "type": "string",
      "value": "[resourceId('Microsoft.Compute/virtualMachines', parameters('vmName'))]"
    }
  }
}
```

### Explicao das Sees do Exemplo

1. **$schema**
   - Define o esquema que valida o template.  uma URL que aponta para o esquema JSON do Azure Resource Manager.

2. **contentVersion**
   - Especifica a verso do template.  usado para controlar a verso do prprio template.

3. **parameters**
   - Define parmetros como `vmName`, `adminUsername` e `adminPassword`, que so valores passados durante a implantao.

4. **variables**
   - Define variveis como `location`, que podem ser usadas em todo o template para simplificar referncias.

5. **resources**
   - Lista os recursos que sero criados. Neste exemplo, estamos criando uma mquina virtual com as propriedades definidas, como perfil de hardware, perfil de sistema operacional, perfil de armazenamento e perfil de rede.

6. **outputs**
   - Define os valores que so retornados aps a implantao do template. Neste exemplo, o `vmId` da mquina virtual criada  retornado.

### Como Implantar um Template ARM

Voc pode implantar um template ARM usando vrias ferramentas:

#### Usando Azure CLI

```sh
az deployment group create --resource-group myResourceGroup --template-file template.json --parameters vmName=myVM adminUsername=azureuser adminPassword=Azure123456!
```

#### Usando Azure PowerShell

```powershell
New-AzResourceGroupDeployment -ResourceGroupName "myResourceGroup" -TemplateFile "template.json" -vmName "myVM" -adminUsername "azureuser" -adminPassword "Azure123456!"
```

#### Usando o Portal do Azure

1. Acesse o Portal do Azure.
2. V para "Grupos de Recursos" e selecione o grupo de recursos onde deseja implantar o template.
3. Clique em "Adicionar" e escolha "Modelos de implantao".
4. Cole o contedo do seu template ARM e clique em "Revisar + criar".

### Prticas Recomendadas

1. **Parmetros e Variveis**
   - Use parmetros para valores que podem mudar entre implantaes e variveis para valores que sero reutilizados dentro do template.

2. **Modularizao**
   - Divida templates complexos em templates menores e reutilizveis. Utilize Linked Templates para compor templates maiores.

3. **Controle de Verso**
   - Armazene templates ARM em um sistema de controle de verso (como Git) para rastrear mudanas e colaborar com a equipe.

4. **Documentao**
   - Adicione metadados e descries para parmetros e recursos no template para facilitar o entendimento e manuteno.

5. **Testes**
   - Teste templates em ambientes de desenvolvimento antes de implantar em produo para garantir que funcionem conforme o esperado.

Os ARM Templates so uma ferramenta poderosa para gerenciar infraestrutura no Azure de forma consistente, automatizada e eficiente.

**Bicep**



Bicep  uma linguagem de domnio especfico (DSL) para a implementao de recursos do Azure, projetada para oferecer uma experincia mais amigvel e menos propensa a erros do que os ARM Templates tradicionais. Bicep simplifica a criao e o gerenciamento de infraestrutura como cdigo no Azure, proporcionando uma sintaxe mais limpa e concisa.

### Principais Caractersticas do Bicep

1. **Sintaxe Simplificada**
   - Bicep oferece uma sintaxe mais simples e limpa em comparao com os ARM Templates JSON, tornando o cdigo mais legvel e fcil de escrever.

2. **Recursos e Funes Integrados**
   - Suporte completo para todos os recursos e funes do ARM, permitindo que voc utilize todas as capacidades do Azure Resource Manager.

3. **Modularidade**
   - Permite criar mdulos reutilizveis, facilitando a manuteno e reutilizao do cdigo.

4. **Validao e Completude de Cdigo**
   - Ferramentas de desenvolvimento, como Visual Studio Code, oferecem suporte a Bicep com recursos como validao em tempo real, completude de cdigo e IntelliSense.

### Exemplo de Template Bicep

Vamos criar um exemplo bsico de um template Bicep que provisiona um grupo de recursos e uma mquina virtual.

#### Cdigo Bicep

```bicep
param vmName string = 'myVM'
param adminUsername string
param adminPassword string
param location string = resourceGroup().location

resource vm 'Microsoft.Compute/virtualMachines@2021-03-01' = {
  name: vmName
  location: location
  properties: {
    hardwareProfile: {
      vmSize: 'Standard_DS1_v2'
    }
    osProfile: {
      computerName: vmName
      adminUsername: adminUsername
      adminPassword: adminPassword
    }
    storageProfile: {
      imageReference: {
        publisher: 'Canonical'
        offer: 'UbuntuServer'
        sku: '18.04-LTS'
        version: 'latest'
      }
      osDisk: {
        createOption: 'FromImage'
      }
    }
    networkProfile: {
      networkInterfaces: [
        {
          id: resourceId('Microsoft.Network/networkInterfaces', 'myNic')
        }
      ]
    }
  }
}

output vmId string = vm.id
```

### Explicao do Exemplo Bicep

1. **param**
   - Define os parmetros que podem ser passados durante a implantao. No exemplo, `vmName`, `adminUsername`, `adminPassword` e `location`.

2. **resource**
   - Define os recursos que sero criados. No exemplo, uma mquina virtual (`Microsoft.Compute/virtualMachines`) com suas propriedades.

3. **output**
   - Define os valores de sada que podem ser retornados aps a implantao. No exemplo, o ID da mquina virtual (`vmId`).

### Implantao de Templates Bicep

Voc pode implantar templates Bicep usando vrias ferramentas:

#### Usando Azure CLI

```sh
az deployment group create --resource-group myResourceGroup --template-file main.bicep --parameters vmName=myVM adminUsername=azureuser adminPassword=Azure123456!
```

#### Usando Azure PowerShell

```powershell
New-AzResourceGroupDeployment -ResourceGroupName "myResourceGroup" -TemplateFile "main.bicep" -vmName "myVM" -adminUsername "azureuser" -adminPassword "Azure123456!"
```

### Converso de ARM Templates para Bicep

Se voc j possui ARM Templates existentes, pode convert-los para Bicep usando a ferramenta `bicep decompile`:

```sh
bicep decompile template.json
```

Isso gera um arquivo `.bicep` equivalente ao template JSON.
